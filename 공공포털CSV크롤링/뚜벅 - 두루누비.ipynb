```python
xml_obj = bs4.BeautifulSoup(contents,'lxml-xml')
rows = xml_obj.findAll('item')
print(rows)
```

```python
columns = rows[0].find_all()
print(columns)
```

```python
# 각 행의 컬럼, 이름, 값을 가지는 리스트 만들기
row_list = [] # 행값
name_list = [] # 열이름값
value_list = [] #데이터값

# xml 안의 데이터 수집
for i in range(0, len(rows)):
    columns = rows[i].find_all()
    #첫째 행 데이터 수집
    for j in range(0,len(columns)):
        if i ==0:
            # 컬럼 이름 값 저장
            name_list.append(columns[j].name)
        # 컬럼의 각 데이터 값 저장
        value_list.append(columns[j].text)
    # 각 행의 value값 전체 저장
    row_list.append(value_list)
    # 데이터 리스트 값 초기화
    value_list=[]
    
#xml값 DataFrame으로 만들기
dorunubi_df = pd.DataFrame(row_list, columns=name_list)
print(dorunubi_df.head(19)) 
```

```python
dorunubi_df = dorunubi_df.drop(columns=['routeIdx','crsIdx','brdDiv','createdtime','modifiedtime'])
```
